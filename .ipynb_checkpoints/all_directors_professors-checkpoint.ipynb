```python
df2 = pd.read_csv('cucat.csv')
```

```python
import math
import numpy as np 
uid, name, verbal, logical, aptitude = [], [], [], [], []
for i, row in df2.iterrows():
    if i > 5:
#         print(row[1], type(row[1]))
        if type(row[1]) != str:
            continue
        print(row[1],row[2], row[3], row[4], row[5], row[6], row[7])
        uid.append(row[1])
        name.append(row[2])
        verbal.append(row[4])
        logical.append(row[5])
        aptitude.append(row[6])
```

```python
data = {
    'uid': uid,
    'name': name,
    'verbal ability': verbal,
    'logical reasoning': logical,
    'quantitative aptitude': aptitude
}

record = pd.DataFrame(data)
```

```python
record.set_index('uid', inplace=True)
```

```python
def get_student_details(id):
#     id= id.upper()
    if isinstance(id, str):
        # If UID is provided
        print(id)
        return record.loc[id]
    else:
        # If Name is provided
        return record[record['name']==id]
```
